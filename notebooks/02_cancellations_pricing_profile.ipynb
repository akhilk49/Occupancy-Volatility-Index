# Day 1 — Cancellations & Seasonal Pricing Profile (Manuel)

**Goal:** Profile `data/raw/cancellations.csv` and `data/raw/seasonal_pricing.csv` — columns, dtypes, nulls, duplicates, date formats.

**Done-when:** These notes committed under `notebooks/`.


```python
from pathlib import Path
import pandas as pd

REPO_ROOT = Path.cwd() if (Path.cwd() / 'data').exists() else Path.cwd().parent
CANCELS_PATH = REPO_ROOT / 'data' / 'raw' / 'cancellations.csv'
PRICING_PATH = REPO_ROOT / 'data' / 'raw' / 'seasonal_pricing.csv'

print(f"Cancellations Exists: {CANCELS_PATH.exists()}")
print(f"Pricing Exists: {PRICING_PATH.exists()}")
```

## 1. Profile Cancellations

```python
if CANCELS_PATH.exists():
    df_cancels = pd.read_csv(CANCELS_PATH)
    print(f"Shape: {df_cancels.shape[0]:,} rows x {df_cancels.shape[1]} columns")
    display(df_cancels.dtypes.to_frame(name='dtype'))
    
    null_summary = pd.DataFrame({
        'null_count': df_cancels.isnull().sum(),
        'null_pct':   (df_cancels.isnull().mean() * 100).round(2),
    })
    print("\nNulls:")
    display(null_summary[null_summary['null_count'] > 0])
    
    print(f"\nExact duplicates: {df_cancels.duplicated().sum()}")
else:
    print("cancellations.csv not found. Waiting for raw data to be populated.")
```

## 2. Profile Seasonal Pricing

```python
if PRICING_PATH.exists():
    df_pricing = pd.read_csv(PRICING_PATH)
    print(f"Shape: {df_pricing.shape[0]:,} rows x {df_pricing.shape[1]} columns")
    display(df_pricing.dtypes.to_frame(name='dtype'))
    
    null_summary = pd.DataFrame({
        'null_count': df_pricing.isnull().sum(),
        'null_pct':   (df_pricing.isnull().mean() * 100).round(2),
    })
    print("\nNulls:")
    display(null_summary[null_summary['null_count'] > 0])
    
    print(f"\nExact duplicates: {df_pricing.duplicated().sum()}")
else:
    print("seasonal_pricing.csv not found. Waiting for raw data to be populated.")
```
